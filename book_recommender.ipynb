### Book Recommendation System using Clustering | Collaborative-based

```python
books = books[['ISBN','Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher','Image-URL-L']]
```

```python
books.rename(columns={
    "Book-Title":"title",
    "Book-Author":"author",
    "Year-Of-Publication":"year",
    "Publisher":"publisher",
    "Image-URL-L":"img_url"}, inplace=True)
```

```python
ratings = pd.read_csv('book_datasets/BX-Book-Ratings.csv', sep=";", on_bad_lines='skip', encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
ratings.rename(columns={
    "User-ID":"user_id",
    "Book-Rating":"rating"}, inplace=True)
```

```python
ratings.head()
```

```python
ratings['user_id'].value_counts()
```

```python
ratings['user_id'].unique().shape
```

```python
x = ratings['user_id'].value_counts() > 200
x[x].shape
```

```python
y = x[x].index
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
books.head()
```

```python
ratings_with_books = ratings.merge(books, on="ISBN")
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
num_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
num_rating.head()
```

```python
num_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
num_rating.head()
```

```python
final_rating = ratings_with_books.merge(num_rating, on='title')
final_rating.head()
```

```python
# Lets take those books which got at least 50 rating of user
final_rating = final_rating[final_rating['num_of_rating'] >= 50]
final_rating.head()
```

```python
final_rating.shape
```

```python
# lets drop the duplicates
final_rating.drop_duplicates(['user_id','title'],inplace=True)
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating.isnull().sum()
```

```python
# Lets create a pivot table
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values= 'rating')
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
model.fit(book_sparse)
```

```python
book_pivot.iloc[237,:]
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
#keeping books name
book_names = book_pivot.index
book_names
```

```python
import pickle
pickle.dump(model, open('components/model.pkl', 'wb'))
pickle.dump(book_names, open('components/book_names.pkl', 'wb'))
pickle.dump(final_rating, open('components/final_rating.pkl', 'wb'))
pickle.dump(book_pivot, open('components/book_pivot.pkl', 'wb'))
```

### Testing the model

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name = "2nd Chance"
recommend_book(book_name)
```
